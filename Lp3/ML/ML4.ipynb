<b>Implement Gradient Descent Algorithm to find local minima of a function.</b><br>Ex: Find local minima of y=(x+3)^2 starting from the point x=2

It is an optimisation algorithm used to find min of a function. We optimise parameters to find local min.

Import dependencies

```python
import numpy as np
import matplotlib.pyplot as plt
```

y = (x+3)^2<br>
After differentiating, y = 2(x+3) = slope

```python
slope = lambda x: (x+3)*2
```

```python
print(slope(2))
```

Initialize terms

```python
x = 2                  
learning_rate = 0.01   
precision = 0.000001
max_it = 500 

prev_step_size = 1 
it = 0 
```

Gradient Descent Algorithm

```python
while prev_step_size > precision and it < max_it:
    prev_x = x 
    
    # New Value = Old Value - Step size 
    # where Step size = learning rate * Slope of function
    x = x - (learning_rate * slope(prev_x)) # Iteration 1: x = 2 - (0.01*10) = 1.9
    
    # Calculate step size
    prev_step_size = abs(x - prev_x)     # Iteration 1: prev_step_size = abs(1.9-2) = 0.1
    
    it = it+1 
    print("It", it, "-", x) 
```

Print Local Minimum

```python
print("\nLocal minimum: ", x)
```

Plot the graph

```python
x_g = np.linspace(-5, -1) # Return evenly spaced numbers
y_g = []

for i in x_g:
  y_g.append((i+3)**2)

plt.plot(x_g,y_g)
```
